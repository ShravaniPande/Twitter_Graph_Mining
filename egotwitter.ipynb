```python
def read_circles(filename):
    circles = []
    un_users = set()
    for line in open(filename):
        lst = line.split('\t')
#         e1, es = lst.split('\t')
        el, es = lst[0], lst[1:]
        circle = set()
        for e in es:
            circle.add(int(e))
            un_users.add(int(e))
        circles.append(circle)
    return circles, un_users
```
