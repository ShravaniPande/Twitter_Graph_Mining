## Plot for algorithm comparison

```python
import matplotlib.pyplot as plt
```

```python
myDictionary = {"Louvain":151.44,"k-Cliques":79.12,"Spectral":109.48,"Girvan-Newman":139.12}
```

```python
fig = plt.figure()
fig.suptitle('Average loss comparison for algorithms', fontsize=14, fontweight='bold')
width = 0.5
ax = fig.add_subplot(111)
# ax.set_title('axes title')
ax.set_xlabel('Algorithms')
ax.set_ylabel('Average Loss')
plt.bar(myDictionary.keys(), myDictionary.values(),width, color='b')
```
